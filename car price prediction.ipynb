```python
q=list(df['Car_Name'])
p=list(set(q))
a=[i for i in range(1,len(p)+1)]
for i in range(len(p)):
    df['Car_Name']=df['Car_Name'].replace(p[i],a[i])
```

```python
df
```

```python
df['Fuel_Type'].value_counts()
```

```python
df['Fuel_Type']=df['Fuel_Type'].replace('Petrol',1)
df['Fuel_Type']=df['Fuel_Type'].replace('Diesel',2)
df['Fuel_Type']=df['Fuel_Type'].replace('CNG',3)
```

```python
df
```

```python
df['Selling_type'].value_counts()
```

```python
df['Selling_type']=df['Selling_type'].replace('Dealer',1)
df['Selling_type']=df['Selling_type'].replace('Individual',0)
```

```python
df
```

```python
df['Transmission'].value_counts()
```

```python
df['Transmission']=df['Transmission'].replace('Manual',0)
df['Transmission']=df['Transmission'].replace('Automatic',1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Selling_Price',axis=1)
```

```python
y=df[['Selling_Price']]
```

```python
x
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=39)
```

```python
m1=LinearRegression()
```

```python
m1.fit(xtrain,ytrain)
```

```python
m1.score(xtrain,ytrain)
```

```python
m1.score(xtest,ytest)
```

```python
m2=DecisionTreeRegressor()
```

```python
m2.fit(xtrain,ytrain)
```

```python
m2.score(xtrain,ytrain)
```

```python
m2.score(xtest,ytest)
```

```python
m3=RandomForestRegressor()
```

```python
m3.fit(xtrain,ytrain)
```

```python
m3.score(xtrain,ytrain)
```

```python
m3.score(xtest,ytest)
```
